```python
sent.rename(columns={'Sentiment' : 'sad or happy'},inplace= True)
sent.head()
```

```python
from sklearn.preprocessing import LabelEncoder
set_ = LabelEncoder()
sent['sad or happy'] = set_.fit_transform(sent['sad or happy'])
```

```python
sent.duplicated().sum()
```

```python
X=sent['SentimentText']
y= sent['sad or happy']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_tcount= cv.fit_transform(X.values)
X_tcount.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_tcount,y)
```

```python
from sklearn.metrics import confusion_matrix , recall_score , precision_score
from sklearn.metrics import accuracy_score
```

```python
model.score(X_tcount,y)
```
